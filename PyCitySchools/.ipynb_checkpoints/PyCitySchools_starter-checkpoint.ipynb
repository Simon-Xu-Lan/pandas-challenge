## Format Function

```python
def currency_format(x):
    """Add $ symbol and comma style to number x, with 2 decimal """
    return "${:,.2f}".format(x)

# Example
df = pd.DataFrame(
        {'A':['A','B','C','D'],
         'C':[123550000000.00,12555.67,640.00,7000]
        })

df['C'] = df['C'].apply(currency_format)
df
```

```python
def comma_format(x):
    """Add comma style to number x, with zero decimal """
    return "{:,.0f}".format(x)

# Example
df = pd.DataFrame(
        {'A':['A','B','C','D'],
         'C':[123550000000.00,12555.67,640.00,7000]
        })

df['C'] = df['C'].apply(currency_format)
df
```

```python
# Percentage format
def percentage_format(x):
    """Add percentage format to number x, with 2 decimal """
    return "{:.2f}%".format(x * 100)

# Example:
df = pd.DataFrame(
        {'A':['A','B','C','D'],
         'C':[123550000000.00,12555.67,640.00,7000]
        })

df['C'] = df['C'].apply(percentage_format)
df
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Create a dictionary 
district_summary = {
    "Total Schools": len(school_data_complete_df.loc[:, "school_name"].unique()),
    "Total Students": len(school_data_complete_df.loc[:, "Student ID"].unique()),
    "Total Budget": school_data_df.loc[:, "budget"].sum(),
    "Average Math Score": school_data_complete_df.mean()["math_score"],
    "Average Reading Score": school_data_complete_df.mean()["reading_score"],  
}

is_passing_math = school_data_complete_df.loc[:,"math_score"] >= 70
students_passing_math = school_data_complete_df[is_passing_math].loc[:, "Student ID"].count()
district_summary["% Passing Math"] = students_passing_math / district_summary["Total Students"]

is_passing_reading = school_data_complete_df.loc[:,"reading_score"] >= 70
students_passing_reading = school_data_complete_df[is_passing_reading].loc[:, "Student ID"].count()
district_summary["% Passing Reading"] = students_passing_reading / district_summary["Total Students"]

students_overall_passing = school_data_complete_df[is_passing_math & is_passing_reading].loc[:, "Student ID"].count()
district_summary["% Overall Passing"] = students_overall_passing / district_summary["Total Students"]

# Create a df
district_summary_df = pd.DataFrame([district_summary])

# Created a formated df
formated_district_summary_df = pd.DataFrame()
formated_district_summary_df["Total Schools"] = district_summary_df["Total Schools"]
formated_district_summary_df["Total Students"] = district_summary_df["Total Students"].apply(comma_format)
formated_district_summary_df["Total Budget"] = district_summary_df["Total Budget"].apply(currency_format)
formated_district_summary_df["Average Math Score"] = round(district_summary_df["Average Math Score"], 2)
formated_district_summary_df["Average Reading Score"] = round(district_summary_df["Average Reading Score"], 2)
formated_district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].apply(percentage_format)
formated_district_summary_df["% Passing reading"] = district_summary_df["% Passing Reading"].apply(percentage_format)
formated_district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].apply(percentage_format)

formated_district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Calculate the average score
groupby_school_data = school_data_complete_df.groupby(["school_name"])
avg_by_school_df = groupby_school_data.mean().loc[:, ["math_score", "reading_score"]]
avg_by_school_df

# Calculate the passing percentage
total_students_per_school  = groupby_school_data.count().loc[:, "Student ID"]
pass_math_sr = school_data_complete_df[is_passing_math].groupby(["school_name"]).count().loc[:, "Student ID"]
percent_pass_math_sr = pass_math_sr / total_students_per_school
pass_reading_sr = school_data_complete_df[is_passing_reading].groupby(["school_name"]).count().loc[:, "Student ID"]
percent_pass_reading_sr = pass_reading_sr / total_students_per_school
pass_overall_sr = school_data_complete_df[is_passing_reading & is_passing_math].groupby(["school_name"])\
                                                                                        .count().loc[:, "Student ID"]
percent_pass_overall_sr = pass_overall_sr / total_students_per_school

# Merge avg_groupby_school_df and school_data_df
school_summary_df = pd.merge(school_data_df, avg_by_school_df, on="school_name")
school_summary_df = school_summary_df.set_index("school_name").sort_index()

school_summary_df = school_summary_df.rename(columns={
    "type": "School Type",
    "size": "Total Students",
    "budget": "Total School Budget",
    "math_score": "Average Math Score",
    "reading_score": "Average Reading Score"
})
school_summary_df["Per Student Budget"] = school_summary_df.loc[:, "Total School Budget"] \
                                            / school_summary_df.loc[:, "Total Students"]

# Reorganize columns
school_summary_df = school_summary_df.loc[:,["School Type", "Total Students", "Total School Budget", \
                                             "Per Student Budget", "Average Math Score", "Average Reading Score"]]


school_summary_df["% Passing Math"] = percent_pass_math_sr
school_summary_df["% Passing Reading"] = percent_pass_reading_sr
school_summary_df["% Overall Passing"] = percent_pass_overall_sr

# Format df
formated_school_summary_df = school_summary_df.copy()
formated_school_summary_df["Total Students"] = school_summary_df.loc[:, "Total Students"].map("{:,.0f}".format)
formated_school_summary_df["Total School Budget"] = school_summary_df.loc[:, "Total School Budget"]\
                                                                                        .map("${:,.0f}".format)
formated_school_summary_df["Per Student Budget"] = school_summary_df.loc[:, "Per Student Budget"]\
                                                                                        .map("${:,.1f}".format)
formated_school_summary_df["Average Math Score"] = school_summary_df.loc[:, "Average Math Score"]\
                                                                                        .map("{:.2f}".format)
formated_school_summary_df["Average Reading Score"] = school_summary_df.loc[:, "Average Reading Score"]\
                                                                                        .map("{:.2f}".format)
formated_school_summary_df["% Passing Math"] = school_summary_df.loc[:, "% Passing Math"]\
                                                                                        .map("{:.2%}".format)
formated_school_summary_df["% Passing Reading"] = school_summary_df.loc[:, "% Passing Reading"]\
                                                                                        .map("{:.2%}".format)
formated_school_summary_df["% Overall Passing"] = school_summary_df.loc[:, "% Overall Passing"]\
                                                                                        .map("{:.2%}".format)
formated_school_summary_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 584.9, 629.9, 644.9, 680]
group_names = ["<$585", "$585-630", "$630-645", "645-680"]
school_summary_df.loc[:, "Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, \
                                                    labels=group_names, include_lowest=True)
score_by_school_spending_df = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()\
    .loc[:, ["Average Math Score", "Average Reading Score", "% Passing Math", \
            "% Passing Reading", "% Overall Passing"]]
score_by_school_spending_df

# Format df
# Create a dictionary to store format requirement for each column
format_requirements = {
    "Average Math Score": "{:.2f}",
    "Average Reading Score": "{:.2f}",
    "% Passing Math": "{:.2%}", 
    "% Passing Reading": "{:.2%}",
    "% Overall Passing": "{:.2%}" 
}
# Ussing for loop to format each column
formated_score_by_school_spending_df = pd.DataFrame()
for column in score_by_school_spending_df.columns:
    formated_score_by_school_spending_df[column] = score_by_school_spending_df.loc[:, column]\
                                                                .map(format_requirements[column].format)

formated_score_by_school_spending_df
```

```python
 def school_performance(**kwargs):
    """
    Create a function to do similar operations:
    1. Create a table that breaks down school performances based on average Spending Ranges (Per Student).
    2. Create a table that breaks down school performances based on school size.
    3. Create a table that breaks down school performances based on school type.
    """
    
    # The dataframe to deal with
    df = kwargs.get("df")
    # The bins to be used in cut method
    bins = kwargs.get("bins")
    # The names for label
    group_names = kwargs.get("group_names")
    # The name of the column to be cut
    by_column = kwargs.get("by_column")
    # New column name to create created
    index_name = kwargs.get("index_name")
    # column names that shall be in new table
    new_columns = kwargs.get("new_columns")
    formats = kwargs.get("formats")
    
    # If by_column is provided, do cut based on by_column
    if by_column != None:
        df.loc[:, index_name] = pd.cut(df[by_column], bins, labels=group_names, include_lowest=True)
    
    # If by_column is not provides, do groupby based on index_name directly.
    new_df = df.groupby([index_name]).mean().loc[:, new_columns]
    

    # Create a formated df
    formated_df = pd.DataFrame()
    for column in new_df.columns:
        formated_df[column] = new_df.loc[:, column].map(formats[column].format)
    
    # Create an empty dictionary and assign it to variable result
    result = {}
    result["df"] = new_df
    result["formated_df"] = formated_df
    
    return result
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0, 999, 1999, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
new_table_columns = ["Average Math Score", "Average Reading Score", "% Passing Math", \
                                                            "% Passing Reading", "% Overall Passing"]

scores_by_school_size = school_performance(
                            df=school_summary_df, 
                            by_column="Total Students", 
                            bins=bins, 
                            group_names=group_names,
                            index_name="School Size",
                            new_columns=new_table_columns,
                            formats=format_requirements
                        )

scores_by_school_size["formated_df"]
```
